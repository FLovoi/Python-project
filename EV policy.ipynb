# Policies to stimulate Electic Vehicle sales
The Netherlands government has implemented various incentives, in order to boost new Electric Vehicle (EV) sales, as a primary option to drastically reduce greenhouse gases emissions (namely CO2). Incentives are categorized into subsidies and tax exemptions. A brief overview of the maximum subsidies is shown below: 
- Used EV car subsidy (for private owners): 2000 Euros 
- New EV car subsidy (for private owners): 2950 Euros 
- New EV car subsidy (for company fleets): 5000 Euros

As far as tax exemptions and reductions are concerned, these are implemented into two tax types: Purchase tax (BPM) and ownership tax (MRB). These are summarized below:
- For BPM:
    - Full exemption until the end of 2024
    - Discount of 360 Euros per car for 2025
    - Discounted, but gradually increasing BPM for 2026 and onwards
- For MRB:
    - Full exemtion until the end of 2024
    - 75% discount for 2025 
    - 50% discount for 2026 and onwards
