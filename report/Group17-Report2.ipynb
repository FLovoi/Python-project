# 2. Policies to stimulate Electic Vehicle sales
The Netherlands government has implemented various incentives, which were put into place on July 2020, in order to boost new Electric Vehicle (EV) sales, as a primary option to drastically reduce greenhouse gases emissions (namely CO2). Incentives are categorized into subsidies and tax exemptions. A brief overview of the maximum subsidies is shown below: 
- Used EV car subsidy (for private owners): 2000 Euros 
- New EV car subsidy (for private owners): 2950 Euros 
- New EV car subsidy (for company fleets): 5000 Euros

As far as tax exemptions and reductions are concerned, these are implemented into two tax types: Purchase tax (BPM) and ownership tax (MRB). These are summarized below:
- For BPM:
    - Full exemption until the end of 2024
    - Discount of 360 Euros per car for 2025
    - Discounted, but gradually increasing BPM for 2026 and onwards
- For MRB:
    - Full exemtion until the end of 2024
    - 75% discount for 2025 
    - 50% discount for 2026 and onwards

For the purposes of this study, we will focus on the 8-year period between 2016 and 2024, which is split into two particular periods: Pre-incentive period (2016-2020) and post-incentive period (2020-2024). Subsequently, comparisons between those two will be conducted.

# 4. Emission Analysis

The implementation of Euro emissions standards has played a crucial role in the reduction of the CO2 emissions of vehicles in Europe. This had huge contribution in improving the air quality and fighting climate change. These standards have been introduced in 1992 with Euro 1. These standards regulated the emissions of new vehicles, and they have become more and more stringent in each subsequent version. In 1996, Euro 2 has been introduced and it reduced the limits for carbon monoxide and the combined limit for unburned hydrocarbons and nitrogen oxide, as well as introducing different levels for petrol and diesel engines. Euro 3, implemented in 2000, and Euro 4 in 2005, brought significant advancement in emission reduction technologies. Later, Euro 5 in 2009 and Euro 6 in 2014 have introduced  stricter regulations that emphasized the fuel efficieny and has promoted the adoption of particulate filters and catalytic converters to reduce the harmful emissions, including CO2. The constant tightening of these standards, combined with the advancement in technology and the widespread of EVs has contributed to a gradual decline in CO2 emissions from the automotive sector.  (Euro 1 to Euro 6 â€“ Vehicle Emissions Standards | RAC Drive, n.d.)

The plot shows CO2 emissions by transport type over time. In particular, the transport types analyzed are Passenger cars, Motorcycles, vehicles with moped licenses, Light-weight commercial motor vehicles, Heavy-weight commercial motor vehicles, and Buses.
The general trend that is possible to analyze is that the highest CO2 emissions are observed in passenger cars, followed by light-weight and heavy-weight, buses, and motorcycles. 
The total emissions of CO2 show an increase from 1990 until around 2005. However, after that, they started to stabilize. Furthermore, around 2020 there was a noticeable dip which is attributed to the decrease in transportation activities related to the Covid 19. The stricter regulations may have reduced the emissions of a single new vehicle, there are still a significant amount of old vehicles on Dutch roads. The percentage of older vehicles in increasing year over year, explaining the gradual impact of the Euro emissions standards (https://www.cbs.nl/en-gb/figures/detail/82044eng). So, to conclude there are two main changes the post-2005 decline and the pandemic-related sharp decline in 2020.


The graph above shows normalized emission levels for some of the most relevant pollutants. A normalized plot is used as the difference in total emissions by weight varies a lot by pollutant. gives insights into the trends of four different pollutants: CO2, N2O, methane (CH4), and NOx from 1990 to 2023.
 
The carbon dioxide (CO2) is possible to identify with the blue line. Its trend starts low in the 1990s and steadily rises until 2004, where it reaches a peak. After this, the CO2 emissions gradually decline, with a significant dip around 2020 due to COVID-19, followed by a rebound. The decline coincides with the implementation of Euro 4, which drastically tightens the limit of emissions.
 
The nitrous oxide (N20) is identified with the red dashed line, and it shows a concerning increase trend. Interesting also to see how the N2O is only partially impacted by COVID. Furthermore, there are no Euro standards that directly want to address this problem, which is also the reason why it keeps increasing; however, it is important to implement policies to lower it since its global warming potential is concerning.
 
The methane (CH4) is indicated with the green line; the trend shows a steep decline from the early 1990s until almost zero by 2020. The Euro standards have become more and more stringent; this has led to improved fuel injection and combustion technologies, which have helped the reduction of methane emissions. Its reduction is a very good sign; however, the maintenance of its low level is essential.
 
Nitrogen Oxides (Nox) are shown in the graph with the yellow dotted line; it has a similar trend to the methane; declining consistently over time. The Euro 3 and Euro 4 have been essential years to tighten the limits, particularly for the diesel vehicles. The Nox limits implemented with Euro 3 for diesel were 0.50 g/km, while for petrol they were 0.15 g/km, and with the implementation of Euro 4, these have been even more restricted. This limit has been decreased until the Euro 6, when they have become 0.18 g/km for diesel and 0.06 g/km for petrol.
 
The different trends among pollutants show how the control of emissions is very complicated, and even if the regulatory measures and technological advancements have reduced CH4 NOx and CO2 +, N2O reduction remain a challenge. However, this is essential in the reduction of the overall environmental and climate impact of transportation emissions.
